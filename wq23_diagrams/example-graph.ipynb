```python
import pandas as pd
```

```python
d = {"stream-64-stride-1.riscv": [43.12, 33.1, 12.1, 1.1],
     "stream-64-stride-4.riscv": [60.12, 32.3, 19.9, 1.2],}
df = pd.DataFrame(d, index=["system1", "system2", "system3", "system4"])
print(df)
```

```python
plot = df.plot.bar(figsize=(10, 5), rot=0)
plot.set_xlabel("systems")
plot.set_ylabel("CPI")
plot.set_title("CPI of Different Systems Running Stream Benchmarks of Different Strides")
print()
```

```python
plot.get_figure().savefig("CPI-graph-example.svg", format='svg', dpi=1200)
```
